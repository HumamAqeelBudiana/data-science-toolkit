1. Create a numpy array of size 2 x 2

```python
import numpy as napy
N1=napy.array([[2,3],[1,8]])
print(N1)
```

2. Create a random numpy array of size 3 x 4

```python
N2=napy.random.random((3,4))
print(N2)
```

3. Create a numpy array of ones and a numpy array of zeroes

```python
N3=napy.ones((2,2))
print(N3)
N4=napy.zeros((2,2))
print(N4)
```

4. Create a random numpy array of size 6x3 and transpose it

```python
N5=napy.random.random((6,3))
print("Original Numpy:")
print(N5)


transpose=N5.T
print("Transposed Numpy")
print(transpose)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
n, p, m = 2, 4, 6

A = napy.random.rand(n, p)
B = napy.random.rand(p, m)

print("Matrix A (n x p):")
print(A)
print("Matrix B (p x m):")
print(B)

C = napy.dot(A, B)
print("Dot Product (A · B) = C (n x m):")
print(C)
```
